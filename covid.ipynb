# Anomaly Detection

## Check anomaly

```python
cases_month = covid.resample("D")["new_cases"].mean()
```

```python
fig,axes = plt.subplots(1,2,figsize=(15,5))
cases_month.plot(legend=False,ax=axes[0])
pd.DataFrame(cases_month).boxplot(ax=axes[1],grid=False)
```

```python
cases_month_diff = cases_month.diff()
```

```python
fig,axes = plt.subplots(1,2,figsize=(15,5))
cases_month_diff.plot(legend=False,ax=axes[0])
pd.DataFrame(cases_month_diff).boxplot(ax=axes[1],grid=False)
```

## Detect Anomaly

```python
def get_outliers_iqr(df, iqr_mul=3):
    """
    Return upper and lower bound of outliers from pandas dataframe based on IQR indicator.
    :param df: Pandas dataFrame
    :param IQR_mul : IQR_mult > 1.5 - normal outliers, IQR_mul > 3 - extreme outliers
    :return lower_outliers, upper outlliers :
    """
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    IQR = Q3 - Q1

    lower_outliers = df < (Q1 - iqr_mul * IQR)
    upper_outliers = df > (Q3 + iqr_mul * IQR)

    return lower_outliers, upper_outliers
```

```python
IQR_multiplier = 1.5 # change it and have fun ;) 
```

```python
lower_outliers, upper_outliers = get_outliers_iqr(cases_month_diff,iqr_mul=IQR_multiplier)
```

```python
fig,axes = plt.subplots(1,2,figsize=(15,5))

##Real series##
cases_month.plot(ax=axes[0],
                 label = "Series",
                 title="Real series",
                 legend = True)


cases_month_outliers = cases_month[(upper_outliers | lower_outliers)]
cases_month_outliers.plot(ax=axes[0],
                             label="Anomalies",
                             title="Real series",
                             legend=True,style=".")
##Differenced series##
cases_month_diff.plot(ax=axes[1],
                      label = "Series",
                      title="Differenced series",
                      legend = True)


cases_month_diff_outliers = cases_month_diff[(upper_outliers|  lower_outliers)]
cases_month_diff_outliers.plot(ax=axes[1],
                                  label="Anomalies",
                                  title="Differenced series",
                                  legend=True,style=".")
```

## Timeseries without anomalies

```python
fig,axes = plt.subplots(1,2,figsize=(15,5))

##Real series##
cases_month.plot(ax=axes[0],
                 label = "Anomalies",
                 title="Real series",
                 legend = True)

cases_month_no_outliers = cases_month[(~upper_outliers & ~lower_outliers)]
cases_month_no_outliers.plot(ax=axes[0],
                             label="Without Anomalies",
                             title="Real series",
                             legend=True)

##Differenced series##
cases_month_diff.plot(ax=axes[1],
                      label = "Anomalies",
                      title="Differenced series",
                      legend = True)


cases_month_diff_no_outliers = cases_month_diff[(~upper_outliers & ~lower_outliers)]
cases_month_diff_no_outliers.plot(ax=axes[1],
                                  label="Without Anomalies",
                                  title="Differenced series",
                                  legend=True)
```
